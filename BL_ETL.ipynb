```python
import pandas as pd
```

```python
#filename and location of csv
csvfile1 = ('resources/FEWS.csv')

#not used
# csvfile2 = ('resources/light-duty-vehicles-2020-08-29.csv')
```

```python
#csv to df
df1 = pd.read_csv(csvfile1)
#checkpoint
df1
```

```python
#get columns
df1.columns
```

```python
#drop unneeded columns for fuel type1 table
df1_fuel=df1.drop(['displ','feScore', 'ghgScore','ghgScoreA', 'highway08','highwayA08', 
                   'hlv', 'hpv', 'lv2', 'lv4', 'phevBlended', 'pv2', 'pv4', 'guzzler',
                   'trans_dscr','createdOn', 'modifiedOn', 'startStop' , 'co2', 'co2A', 
                   'city08', 'cityA08', 'comb08','combA08', 'UCity', 'UCityA', 
                   'UHighway', 'UHighwayA', 'tCharger', 'sCharger','VClass','mpgData','mfrCode'], axis=1)
```

```python
#determine different fuel types, to split into tables
# df1_fuel['fuelType'].unique()
```

```python
#new columns
df1_fuel.columns
```

```python
#reorder columns
df1_reorder = df1_fuel[['id', 'make', 'model', 'year', 'atvType', 
                        'trany', 
                        ##petroleum
                        'fuelType', 'fuelType1', 'fuelType2',
                        'fuelCost08', 'fuelCostA08', 
                        'barrels08', 'barrelsA08', 
                        ##petroleum measurements
                        'city08U', 'cityA08U', 
                        'highway08U', 'highwayA08U',
                        'comb08U', 'combA08U',
                        'cityCD', 'cityE', 'cityUF',
                        'highwayCD', 'highwayE', 'highwayUF',
                        'combinedCD', 'combE', 'combinedUF',
                        'co2TailpipeAGpm', 'co2TailpipeGpm', 
                        'cylinders', 'drive', 'engId', 'eng_dscr',
                         ##electric
                        'evMotor',
                        'charge120', 'c240Dscr',  'charge240', 
                        'c240bDscr', 'charge240b',
                        ##electric measurements
                        'range','rangeA', 
                        'rangeCity', 'rangeCityA', 
                        'rangeHwy', 'rangeHwyA', 
                        'phevCity', 'phevHwy', 
                        'phevComb',
                        ##evaluation
                        'youSaveSpend'
                       ]
                      ]
```

```python
#make the slice a df
df_prep = pd.DataFrame(df1_reorder)
```

```python
##renaming columns, friendly names, from documentation
df_prep.rename(columns ={'atvType':'advancement_type', 
                        'trany':'transmission',
                        ##petroleum
                        'fuelType':'fuel_types', 'fuelType1':'fuel1', 'fuelType2':'fuel2',
                        'fuelCost08':'fuel1_cost', 'fuelCostA08':'fuel2_cost', 
                        'barrels08': 'fuel1_consumption', 'barrelsA08':'fuel2_consumption', 
                        ##petroleum measurements
                        'city08U': 'fuel1_city_MPG', 'cityA08U':'fuel2_city_MPG', 
                        'highway08U':'fuel1_highway_MPG', 'highwayA08U':'fuel2_highway_MPG',
                        'comb08U': 'fuel1_combined_MPG', 'combA08U': 'fuel2_combined_MPG',
                        'cityCD':'city_CD', 'cityE':'city_E', 'cityUF':'city_UF',
                        'highwayCD':'highway_CD', 'highwayE':'highway_E', 'highwayUF':'highway_UF',
                        'combinedCD':'combined_CD', 'combE': 'combined_E', 'combinedUF':'combined_UF',
                        'co2TailpipeAGpm':'fuel1_co2_GPM', 'co2TailpipeGpm':'fuel2_co2_GPM', 
                        'drive':'drive_axel', 'engId':'engine_ID', 'eng_dscr':'engine_dscr',
                         ##electric
                        'evMotor':'electric_motor', 
                        'charge120':'charge_120v', 'c240Dscr':'charger_dscr','charge240': 'charge_240v', 
                        'c240bDscr':'alternate_charger_dscr','charge240b':'alternate_charger_240v',
                        ##electric measurements
                        'range':'fuel1_range','rangeA':'fuel2_range', 
                        'rangeCity':'fuel1_city_MPG', 'rangeCityA':'fuel2_city_MPG', 
                        'rangeHwy':'fuel1_highway_MPG', 'rangeHwyA':'fuel2_city_MPG', 
                        'phevCity':'hybrid_city_MPG', 'phevHwy':'hybrid_highway_MPG', 
                        'phevComb':'hybrid_combined_MPG',
                        ##evaluation
                        'youSaveSpend': 'save_to_spend'
                       }, inplace=True)
```

```python
#check and rename
df2 = df_prep
df2
```

```python
#filter out cars made before year 2000
df_ev = df2[df2['year'] >= 2000]
```
