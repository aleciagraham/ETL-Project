```python
#fuel databases need to combine 
import pandas as pd
import os
import numpy as np
```

```python
#Import Data
gas = ('resources/gas_prices.csv')
otherfuels = ('resources/Alternative_fuel_prices.csv')
#Reading CSV file
gas = pd.read_csv(gas)
otherfuel = pd.read_csv(otherfuels)
```

```python
#data cleaning to pull out year 2000 to current
gas_db= gas[gas['Year']>1999]
gas_db
```

```python
otherfuel.columns
```

```python
#headers from auto Array(['Regular', 'Premium', 'Diesel', 'CNG', 'Gasoline or natural gas',
  #     'Gasoline or E85', 'Electricity', 'Gasoline or propane',
#       'Premium or E85', 'Midgrade', 'Premium Gas or Electricity',
 #      'Regular Gas and Electricity', 'Premium and Electricity',
  #     'Regular Gas or Electricity'], dtype=object)
```

```python
#merge gasoline w/other fuel dataset
all_fuels = pd.merge(gas_db,otherfuel, on ="Year")
all_fuels
```

```python
#drop columns

all_fuels_db=all_fuels.drop(['Gasoline Price ($/gallon)','LNG','Propane','B2/B5', 'Unnamed: 11', 'Unnamed: 12',
       'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16',
       'Unnamed: 17'], axis = 1) 
all_fuels_db.dtypes
```

```python
#Removed the $ signs to convert data types to float
all_fuels_db['Gasoline'] = all_fuels_db['Gasoline'].str.replace('$','')
all_fuels_db['E85'] = all_fuels_db['E85'].str.replace('$','')
all_fuels_db['CNG'] = all_fuels_db['CNG'].str.replace('$','')
all_fuels_db['Diesel'] = all_fuels_db['Diesel'].str.replace('$','')
all_fuels_db['B20'] = all_fuels_db['B20'].str.replace('$','')
all_fuels_db['B99/B100'] = all_fuels_db['B99/B100'].str.replace('$','')
all_fuels_db['Electricity**'] = all_fuels_db['Electricity**'].str.replace('$','')
```

```python
float_data = all_fuels_db
```

```python
#convert data type from object to float
float_data['Gasoline'] = float_data['Gasoline'].astype(float)
float_data['E85'] = float_data['E85'].astype(float)
float_data['CNG'] = float_data['CNG'].astype(float)
float_data['Diesel'] = float_data['Diesel'].astype(float)
float_data['B20'] = float_data['B20'].astype(float)
float_data['B99/B100'] = float_data['B99/B100'].astype(float)
float_data['Electricity**'] = float_data['Electricity**'].astype(float)
final_data=float_data
```

```python
#to csv file
final_data.to_csv('resources/fuel.csv')
```
